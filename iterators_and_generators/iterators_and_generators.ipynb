## Iterable (Short Note)

An **iterable** is any object that can be looped over (iterated).

It means:
> The object can return its elements one by one.

---

### Examples of Iterables
- List → `[1, 2, 3]`
- Tuple → `(1, 2, 3)`
- String → `"hello"`
- Dictionary → `{"a": 1, "b": 2}`
- Set → `{1, 2, 3}`

---

### How It Works

An iterable can be converted into an iterator using:

```python
iter(obj)
One-line memory trick

👉 Iterable = collection
👉 Iterator = value provider

````python
nums = [1, 2, 3, 4]   # iteratable

# to see the list of dunder methods available
print(dir(nums))
````

````python
# to make the iterable iterator we use
i_nums = nums.__iter__()    # or its equivalent is iter(nums), it calls the iter dunder method in background
i_nums
````

````python
# Iterator is an object with a state that remembers where it is during iteration
print(next(i_nums))
print(next((i_nums)))
print(next((i_nums)))
print(next((i_nums)))
print(next((i_nums)))  # will give StopIteration
````
