### Classes and Instances


```python
## Defining a Class and passing empty class to an instance
class Employee:
    pass

emp1 = Employee()
print(emp1)

## assigning value from instance itself
emp1.name = "anish"
emp1.designation = "AI software engineer"
emp1.email = "anish.upreti@company.com"

print(emp1.name, emp1.designation, emp1.email)
```

```python
class Employee():
    ## initializing or constructor
    def __init__(self, first, last, email):   # self is the instance argument that is automatically passed, it can be named anything other than self, but it is standard practice to use self
        self.first = first # we can use anything in place of .first (.fname,etc)
        self.last = last
        self.email = email
    
    def fullname(self):
        return '{} {}'.format(self.first, self.last)
    
emp1 = Employee("anish", "upreti", "123@company.com")
print(emp1.email)

#calling class function from instance
print(emp1.fullname())

# emp1.fullname() is same as:
print(Employee.fullname(emp1))
```

### Class variables
It is shared across all instances of that class.
It can be accessed with help of class itself or instance of that class.

### Difference Between Class Variable and Instance Variable

A **class variable** is shared by all objects (instances) of a class. It is defined inside the class but outside any method. This means every instance sees the same value unless it is specifically overridden.

Example:
- `interest_rate = 0.05` → all bank accounts use the same interest rate.

An **instance variable** is unique to each object. It is defined inside the `__init__` method using `self`. Each instance can have different values.

Example:
- `self.balance` → each account has its own balance.

### Key Difference:
- Class variable → **shared across all objects**
- Instance variable → **unique per object**

### Simple Analogy:
Think of a class as a template for bank accounts:
- Class variable → like a **common rule** (same interest rate for all accounts)
- Instance variable → like **personal data** (each person’s account balance)

```python
class Employee:
    
    # class variable
    num_of_employees = 0
    raise_rate = 1.04

    def __init__(self, first, last, pay):
        self.first = first
        self.last = last
        self.pay = pay

        # accessing class variable with help of class itself
        Employee.num_of_employees += 1

    def apply_raise(self):
        self.pay = self.pay * self.raise_rate # accessing raise rate class variable through instance

```

```python
emp1 = Employee("anish", "upreti", 100000)
print(emp1.pay)

emp1.apply_raise()

print(emp1.pay)

print(emp1.raise_rate)
print(Employee.raise_rate)
print(Employee.num_of_employees)
```

```python
emp1.raise_rate = 1.05  # this over rides the class variable value for this specific instance

emp1.apply_raise()

print(emp1.pay)
print(emp1.raise_rate)

```

### Q. Solve the below problem.


```python
# Add these to your existing BankAccount class:

# 1. A class variable — interest_rate = 0.05
#    (shared across ALL accounts, not per instance)

# 2. A method — apply_interest(self)
#    adds interest to the current balance

# 3. A class method — set_interest_rate(cls, rate)
#    changes the rate for ALL accounts at once

class BankAccount:
    
    # 1. Class variable (shared by ALL instances)
    interest_rate = 0.05

    def __init__(self, name, balance):
        # Instance variables (unique for each object)
        self.name = name
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        return self.balance

    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
            return self.balance
        else:
            return("Insufficient funds")
    
    def get_balance(self):
        # Returns current balance
        return self.balance

    def __str__(self):
        # String representation of object (used in print)
        return f"Account[{self.name}] — Balance: ${self.balance}"
    
    def apply_interest(self):
        # 2. Add interest to the current balance
        # Uses class variable (interest_rate)
        self.balance = self.balance + (self.balance * self.interest_rate)

    @classmethod
    def set_interest_rate(cls, rate):
        # 3. Class method to update interest rate for ALL accounts
        # cls refers to the class (BankAccount)
        cls.interest_rate = rate


# Create two account objects
acc1 = BankAccount("ram", 1000)
acc2 = BankAccount("hari", 2000)

# Apply interest using default rate
acc1.apply_interest()
print(acc1.get_balance())

# Change interest rate for ALL accounts
BankAccount.set_interest_rate(0.10)

# Apply new interest rate
acc1.apply_interest()
print(acc1.get_balance()) 

acc2.apply_interest()
print(acc2.get_balance())   

acc2.apply_interest()
print(acc2.get_balance())
```

### Class methods and static methods
Regular methods in a class automatically take the instance as the first argument.  
To take class as first argument we use class methods.(use @classmethod)  
Static methods do not pass anything automatically unlike regular and class methods. (use @staticmethod)


```python
class Employee:
    
    # class variable
    num_of_employees = 0
    raise_rate = 1.04

    def __init__(self, first, last, pay):
        self.first = first
        self.last = last
        self.pay = pay

        # accessing class variable with help of class itself
        Employee.num_of_employees += 1

    def apply_raise(self):
        self.pay = self.pay * self.raise_rate # accessing raise rate class variable through instance

    # using class method lets us use the class variable in the function
    @classmethod
    def from_string(cls, emp_str):
        first, last, pay = emp_str.split("-")
        return cls(first, last, pay)
    
    # using static method lets us use the function without passing in instance or class argument
    def is_workday(day):
        if day.weekday() == 5 or day.weekday() == 6:
            return False
        return True
```

```python
emp_str_1 = "anish-upreti-600000"

# using class method
new_emp1 = Employee.from_string(emp_str_1)
print(new_emp1.first)
print(new_emp1.last)
```

```python
import datetime
emp_date = datetime.date(2026, 3, 25)
emp_date2 = datetime.date(2026, 3, 28)

#using static method
print(Employee.is_workday(emp_date))
print(Employee.is_workday(emp_date2))
```

### When to use classmethod vs staticmethod vs regular method

A **regular method** (uses `self`) is used when you need to work with instance data, like updating a specific account’s balance.  

A **classmethod** (uses `cls`) is used when you want to work with or modify class-level data that affects all instances, like changing the interest rate.  

A **staticmethod** is used when the method doesn’t need access to either instance or class data, and is just a utility function related to the class.

### Q. Solve the below problem.


```python
# Add to your existing BankAccount class:

# 1. A classmethod — from_string(cls, account_str)
#    takes a string like "Alice-1000" and creates
#    a BankAccount instance from it

# 2. A classmethod — from_dict(cls, data)
#    takes a dict like {"owner": "Bob", "balance": 500}
#    and creates a BankAccount instance

# 3. A staticmethod — validate_amount(amount)
#    returns True if amount > 0, False otherwise
#    use this inside deposit() and withdraw()
#    to validate before doing anything

class BankAccount:
    
    # Class variable (shared by all instances)
    interest_rate = 0.05

    def __init__(self, name, balance):
        # Instance variables
        self.name = name
        self.balance = balance

    def deposit(self, amount):
        # Use staticmethod to validate amount first
        if BankAccount.validate_amount(amount):
            self.balance += amount
            return self.balance
        return "Invalid amount"

    def withdraw(self, amount):
        # Validate amount first
        if not BankAccount.validate_amount(amount):
            return "Invalid amount"
        
        # Then check if enough balance is available
        if self.balance >= amount:
            self.balance -= amount
            return self.balance
        else:
            return "Insufficient funds"
    
    def get_balance(self):
        # Returns current balance
        return self.balance

    def __str__(self):
        # String representation of object
        return f"Account[{self.name}] — Balance: ${self.balance}"
    
    def apply_interest(self):
        # Adds interest to the current balance
        self.balance = self.balance + (self.balance * self.interest_rate)

    @classmethod
    def set_interest_rate(cls, rate):
        # Changes interest rate for all accounts
        cls.interest_rate = rate

    @classmethod
    def from_string(cls, acc_str):
        # Create object from string like "Alice-1000"
        name, balance = acc_str.split('-')
        return cls(name, float(balance))
    
    @classmethod
    def from_dict(cls, acc_dict):
        # Create object from dictionary like {"owner": "Bob", "balance": 500}
        name = acc_dict["owner"]
        balance = acc_dict["balance"]
        return cls(name, balance)
    
    @staticmethod
    def validate_amount(amount):
        # Returns True only if amount is greater than 0
        return amount > 0
    
```

```python
# testing the above class
acc1 = BankAccount.from_string("Alice-1000")
acc2 = BankAccount.from_dict({"owner": "Bob", "balance": 500})
print(acc1)
print(acc2)
print(BankAccount.validate_amount(-50))  # False
print(BankAccount.validate_amount(200))
```
